## **Reading data**

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

# reading only the first 200 movies
movies_sample = movies.iloc[:200]

# reading only the first 200 users
User_sample = ratings[ratings['userId'] < 201]
#User_sample = User_sample.reset_index(drop=True)

```

## **Getting rid of unwatched movies & combining data**

```python
# merging between movies and ratings based on the movies watched in both files
#combined_movies_data = pd.merge(User_sample, movies_sample, on='movieId')

combined_movies_data = pd.merge(User_sample, movies_sample, on='movieId')

# dropping times_stamps and genres columns
combined_movies_data = combined_movies_data.drop(
    ['timestamp', 'genres'], axis=1)


combined_movies_data
```

## **Creating pivot table (CF tabel)**

```python
# creating the CF tabel
rating_crosstab = combined_movies_data.pivot_table(
    values='rating', index='userId', columns='title', fill_value=0)
print("the Pivot_table size = (userIDs , movie titles) = ", rating_crosstab.shape)
rating_crosstab.head()
```

## **Adjusting the table for item_to_item similarity**

```python
# transposing it to calculate the similarity matrix of item to item
user_ratings_to_movies = rating_crosstab.transpose()
print("transpose Pivot_tabel size = (movie titles , userIDs) = ",
      user_ratings_to_movies.shape)
user_ratings_to_movies.head()
```

## **Creating similarity matrix for the movies**

```python
# importing the cosine_similarity function
from sklearn.metrics.pairwise import cosine_similarity

# calculating the cosine_similarity item to item
similarity_scores_movies = cosine_similarity(user_ratings_to_movies)
print("shape of the matrix is", similarity_scores_movies.shape)
print(similarity_scores_movies)
```

## **Saving the similarity values into a DataFrame**

```python
# creating df with the movies titles and similarity
data_matrix_item = pd.DataFrame(
    data=similarity_scores_movies, index=user_ratings_to_movies.transpose().columns, columns=user_ratings_to_movies.transpose().columns)

# data_matrix_item.head()
data_matrix_item.shape
```

## **Movie to Movie recomender**

```python
# function that takes the name of the movie and the number of recommendations
def recommend_movies_similar_to_movie(movie, recomemnds=10):

    # catching the name of the movie and number of recommendations
    movie = movie
    number_of_recommendation = recomemnds

    # in data_matrix movie titles were set to be the indices
    # saving the indices means savng the mavie titles
    indices = data_matrix_item.loc[movie].nlargest(
        number_of_recommendation+1).index

    print("The recommended movies if u like '{}' are \n".format(movie))
    # looping over the indices and printing the recommended movies
    for idx, num in enumerate(indices):
        if idx > 0:
            print(num)
```

## **Recommending the movies similar to 'Toy Story (1995)'**

```python
# recommending movies similar to ''Toy Story (1995)''
recommend_movies_similar_to_movie('Toy Story (1995)')
```

## **Creating User Similarity Matrix**

```python
# calculate the similarity matrix for the users
similarity_scores_user = cosine_similarity(user_ratings_to_movies.transpose())
similarity_scores_user.shape

# saving the unique userIDS
user_ids = np.unique(combined_movies_data["userId"])

# creating a dataframe where
# similarity matrix for useres is the data
# columns and indices are the user IDS
data_matrix_user = pd.DataFrame(
    data=similarity_scores_user, index=user_ids, columns=user_ids)
```

## **Fuction that retuns the Recommended movies for user**

```python
def recommend_movies_for_user(userID, recommends=3, similar_users=5):

    #catching function args
    userID = userID
    number_of_similar_users = similar_users
    number_of_recommendation = recommends

    #counter to stop recommending
    counter = 0

    # save the ids of similar users
    indices = data_matrix_user.loc[userID].nlargest(number_of_similar_users+1).index

    print("the recommended movies for user {}".format(userID))

    # loop on those similar users
    #print(indices)
    
    for idx, num in enumerate(indices):
        # save all the movies that this user watched with high ratings
        movie_rec_user = rating_crosstab.loc[indices[idx]].nlargest(
            number_of_recommendation+1).index

        # lopp over those movies with high ratings
        for idx, num in enumerate(movie_rec_user):

            # check if the user have watched it or not (if user rated that movie or not)
            if (rating_crosstab[movie_rec_user[idx]].loc[userID] == 0) & (counter < number_of_recommendation):

                # if he haven't watched it recommend the movie
                print(movie_rec_user[idx])
                counter += 1
```
